```python
def drop_rv_feature(data):
    df=data
    """to_drop =['RAW_TP9', 'RAW_AF7', 'RAW_AF8', 'RAW_TP10', 'AUX_RIGHT',
           'Accelerometer_X', 'Accelerometer_Y', 'Accelerometer_Z', 'Gyro_X',
           'Gyro_Y', 'Gyro_Z', 'HeadBandOn', 'HSI_TP9', 'HSI_AF7', 'HSI_AF8',
           'HSI_TP10', 'Battery', 'Elements']"""
    to_drop =['HRV_SDANN1', 'HRV_SDNNI1', 'HRV_SDANN2', 'HRV_SDNNI2',
           'HRV_SDANN5', 'HRV_SDNNI5', 'HRV_ULF','HRV_VLF']
    # 丢弃特征 drop columns
    df.drop(to_drop, axis=1, inplace=True)
    #print("feature droped!")
    return df
def drop_gaze_feature(data):
    df=data

    to_drop =['index','Time']
    # 丢弃特征 drop columns
    df.drop(to_drop, axis=1, inplace=True)
    #print("feature droped!")
    return df
def timestr2datetime(timestr,pattern):
  time_ = time.strptime(timestr, pattern)#to Time object
  mktime_ = time.mktime(time_)#to timestamp
  datetime_ = datetime.datetime.fromtimestamp(mktime_)#to Datetime object
  return datetime_

def savecsv(filepath,data):
  data.to_csv(filepath,index=0)
  print("saved to ",filepath)
  print("-"*50)  

def get_new_columns(df):
    new_columns=[]
    for j in range(len(df.columns)):
      for i in range(len(df.index)):
        column = df.columns[j]
        new_column = column + "_" + df.index[i]
        new_columns.append(new_column)
    return new_columns
```

```python
base_path = "/content/drive/MyDrive/Emotion-RST/vrchat_emotion_data2022/MID_DATA"
year = "2022"
date = "10-19"
name = "Jiang"
dir_name = year+"-"+date+"-"+name #2022-10-12-Yushan
SR_file_name = name+".csv"
```

```python
#SR_filepath = "/content/drive/MyDrive/Emotion-RST/vrchat_emotion_data2022/MID_DATA/2022-10-12-Yushan/Yushan.csv"
SR_filepath = base_path +"/" + dir_name + "/" + SR_file_name
#SR_df = pd.read_csv(SR_filepath, encoding='CP932')
SR_df = pd.read_csv(SR_filepath)
SR_df.columns = ['Time','Answerer','VR world','Topic','Collaborators(Main&aux)','StartTime','EndTime','Valence','Arousal','Dominance','Excite-Not Excite','Serious-Not Serious','Smooth-Not Smooth','Bright-Not Bright','Eminent-Not Eminent','Close-Not Close','Equal-Not Equal']
#SR_df


one_person_gaze_rv_instances_list=[]
#len(SR_df)
for i in range(len(SR_df)):

  start_time_str = SR_df['StartTime'].iloc[i]
  end_time_str = SR_df['EndTime'].iloc[i]

  start_time_str = year+"-"+date+ "_" + start_time_str
  end_time_str = year+"-"+date+ "_" + end_time_str
  #print(start_time_str,end_time_str)
  #print(type(start_time_str))
  pattern = "%Y-%m-%d_%H:%M:%S"
  start_datetime = timestr2datetime(start_time_str,pattern) 
  end_datetime = timestr2datetime(end_time_str,pattern) 
  delta_minute = datetime.timedelta(minutes=1)
  starttime = start_datetime
  endtime = end_datetime
  print("Experiment period:",starttime,endtime)
  print("-"*30)


  #gaze_filepath="/content/drive/MyDrive/Emotion-RST/vrchat_emotion_data2022/MID_DATA/2022-10-12-Yushan/gaze"+str(i)+".csv"
  gaze_filepath = base_path +"/" + dir_name + "/"+ "gaze"+str(i)+".csv"
  gaze_df = pd.read_csv(gaze_filepath)
  #print(gaze_df)

  #rv_filepath="/content/drive/MyDrive/Emotion-RST/vrchat_emotion_data2022/MID_DATA/2022-10-12-Yushan/rv"+str(i)+".csv"
  rv_filepath = base_path +"/" + dir_name + "/"+ "rv"+str(i)+".csv"
  rv_df = pd.read_csv(rv_filepath)
  rv_df = drop_rv_feature(rv_df)
  print(rv_df)


 
  delta_second = datetime.timedelta(seconds=1)
  window_starttime=starttime

  gaze_instances_list=[]
  rv_instances_list=[]
  gaze_rv_instances_list=[]

  round=0
  while(True):

    window_starttime = starttime + round*delta_second
    window_endttime = window_starttime + 15*delta_second
    print(window_starttime,window_endttime)
    gaze_df['Time'] = pd.to_datetime(gaze_df['Time'] )
    gaze_instance_data = gaze_df[(gaze_df['Time'] >=window_starttime) & (gaze_df['Time'] <= window_endttime)]
    gaze_instance_data = drop_gaze_feature(gaze_instance_data)#delete index Time

    rv_df['Time'] = pd.to_datetime(rv_df['Time'] )
    rv_instance_data = rv_df[(rv_df['Time'] >=window_starttime) & (rv_df['Time'] <= window_endttime)]
    rv_instance_data.drop(['Time'], axis=1, inplace=True)#delete Time

    #print(" cut gaze",gaze_instance_data)
    #print(" cut rv",rv_instance_data)
    #print(gaze_instance_data)


    
    #print(gaze_instance_data.columns)
    statistics_gaze_instance_data = gaze_instance_data.describe()
    statistics_gaze_instance_data = statistics_gaze_instance_data.iloc[1:]
    #print(statistics_gaze_instance_data)


    #print(rv_instance_data)
    statistics_rv_instance_data = rv_instance_data.describe()
    statistics_rv_instance_data = statistics_rv_instance_data.iloc[1:]


    new_gaze_columns = get_new_columns(statistics_gaze_instance_data)
 
    new_rv_columns = get_new_columns(statistics_rv_instance_data)

    

    statistics_gaze_instance_data = statistics_gaze_instance_data.to_numpy().T.reshape(1,-1)
    new_statistics_gaze_instance_data = pd.DataFrame(statistics_gaze_instance_data, columns = new_gaze_columns)
    

    statistics_rv_instance_data = statistics_rv_instance_data.to_numpy().T.reshape(1,-1)
    new_statistics_rv_instance_data = pd.DataFrame(statistics_rv_instance_data, columns = new_rv_columns)
    
    #print(new_statistics_gaze_instance_data)

    #print(new_statistics_gaze_instance_data)
    #print(new_statistics_rv_instance_data)

    #self_report_df = SR_df[['Valence','Arousal','Dominance','Excite-Not Excite','Serious-Not Serious','Smooth-Not Smooth','Bright-Not Bright','Eminent-Not Eminent','Close-Not Close','Equal-Not Equal']].iloc[[i]]
    self_report_df = SR_df.iloc[[i]]
    print(self_report_df)
    #input()

    new_statistics_gaze_instance_data = new_statistics_gaze_instance_data.reset_index(drop=True)
    new_statistics_rv_instance_data = new_statistics_rv_instance_data.reset_index(drop=True)
    self_report_df = self_report_df.reset_index(drop=True)

    gaze_instance = pd.concat([new_statistics_gaze_instance_data, self_report_df],axis=1)
    rv_instance = pd.concat([new_statistics_rv_instance_data, self_report_df],axis=1)
    gaze_rv_instance = pd.concat([new_statistics_gaze_instance_data,new_statistics_rv_instance_data, self_report_df],axis=1)

    print(gaze_instance)
    print(rv_instance)
    print(gaze_rv_instance)
    #input()

    gaze_instances_list.append(gaze_instance)
    rv_instances_list.append(rv_instance)
    gaze_rv_instances_list.append(gaze_rv_instance)
    
    round=round+1
    if(window_endttime>endtime):
      break

  gaze_instances = pd.concat(gaze_instances_list)
  rv_instances = pd.concat(rv_instances_list)
  gaze_rv_instances = pd.concat(gaze_rv_instances_list)

  print(gaze_instances)
  print(rv_instances)
  print(gaze_rv_instances)

  gaze_save_filePath = base_path +"/" + dir_name + "/"+ "processed_gaze_instance"+str(i)+".csv"
  savecsv(gaze_save_filePath , gaze_instances) 

  rv_save_filePath = base_path +"/" + dir_name + "/"+ "processed_rv_instance"+str(i)+".csv"
  savecsv(rv_save_filePath , rv_instances) 

  gaze_rv_save_filePath = base_path +"/" + dir_name + "/"+ "processed_gaze_rv_instance"+str(i)+".csv"
  savecsv(rv_save_filePath , gaze_rv_instances) 

  one_person_gaze_rv_instances_list.append(gaze_rv_instances)


one_person_gaze_rv_instances = pd.concat(one_person_gaze_rv_instances_list)
gaze_rv_concated_save_filePath = base_path +"/" + dir_name + "/"+ "processed_gaze_rv_concated.csv"
savecsv(gaze_rv_concated_save_filePath , one_person_gaze_rv_instances) 

      

```
